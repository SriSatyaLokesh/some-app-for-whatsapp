```
from datetime import datetime
import json
def isDateObject(s):
  
  try:
    date_time = datetime.strptime(s, "%d/%m/%Y, %I:%M %p")
    # print(date_time)
    return True
  except:
    return False
chat=[]
import re #regEx
k = 0
with open("WhatsApp Chat with CSE 2K16-2K20.txt","r") as f:
  for line in f:
    
    s = line.split(" - ")
    # print(s)
    if len(s)>0:
      if isDateObject(s[0]):
        message = dict()
        message["datetime"] = s[0]
        k+=1
        user_msg = line.replace(s[0]+" - ","").split(": ")
        user = user_msg[0]
        actions = ["added", "changed" , "removed", "joined","left","created"]
        for i in actions:
          if i in user:
            message["isAction"] = True
            message["action"] = user
            break
        if "isAction" not in message:
            message["user"] = user
            if "<Media omitted>" in user_msg[1]:
              message["isMedia"] = True
            else:
              message["isMessage"]= True
              message["message"] = user_msg[1]
        chat.append(message)
        
      else:
        try:
          l = len(chat) - 1
          chat[l]["message"] = chat[l]["message"]+line
        except Exception as err:
          # print(chat[len(chat)-1])
          print(err)
          print(chat[l])
          print(line)
    
# print(chat) 
chat_json = dict()    
for i in range(len(chat)):
  chat_json[str(i)] = chat[i]

chat_json = json.dumps(chat_json)
print(chat_json)
```
